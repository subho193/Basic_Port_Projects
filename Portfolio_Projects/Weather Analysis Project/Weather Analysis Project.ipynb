### 3. Exploratory Data Analysis(EDA)

> There is good amount of correlation between the Temp_C and Dew Point Temp_C feature..

> ***Most of the features are numerical. So, Let's plot different Distribution plots first..***

```python
###Setting up the background...
sns.set_style('darkgrid')
plt.figure(figsize=(12,8))
sns.histplot(df['Temp_C'])
```

```python
### Let's check the relation between Temperature and Dew Point...
plt.figure(figsize=(12,8))
sns.scatterplot(df['Temp_C'],df['Dew Point Temp_C'],hue=df['Weather'])
plt.legend(bbox_to_anchor=[0.8, 0.5])### Customizing the position of the Legend..
```

***When Temperature increases the Dew also increases...***

```python
plt.figure(figsize=(12,8))
sns.jointplot(df['Temp_C'],df['Dew Point Temp_C'])
plt.legend(bbox_to_anchor=[0.9, 0.5])
plt.show()
```

```python
df['Weather'].value_counts()[df['Weather'].value_counts()>10].index
```

```python
weather=['Mainly Clear', 'Mostly Cloudy', 'Cloudy', 'Clear', 'Snow', 'Rain',
       'Rain Showers', 'Fog', 'Rain,Fog', 'Drizzle,Fog', 'Snow Showers',
       'Drizzle', 'Snow,Fog', 'Snow,Blowing Snow', 'Rain,Snow',
       'Thunderstorms,Rain Showers', 'Haze', 'Drizzle,Snow,Fog',
       'Freezing Rain', 'Freezing Drizzle,Snow']
```

```python
ndf=df[df['Weather'].apply(lambda a: a in weather)]
```

```python
### Now let's plot categorical plots..
plt.figure(figsize=(12,8))
sns.barplot(ndf['Weather'],ndf['Temp_C'])
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(ndf['Weather'],ndf['Temp_C'])
plt.xticks(rotation=45)
plt.show()
```

#### Conclusion : 
- We have taken only those weather categories which contains above 10 numbers of observations and have not considered those weather categories which contains less than or equal to 10 observations..
- When the weather is mainly clear and clear the temperature raises the most and during snow fall the temperature falls the most..
