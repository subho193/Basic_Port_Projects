```python
def age_grouper(a):
    if a>=18 and a<25:
        return '18-25'
    elif a>=25 and a<40:
        return '25-40'
    elif a>=40 and a<55:
        return '40-55'
    elif a>=55:
        return '55+'
```

```python
df['age_group']=df['new_age'].apply(age_grouper)
```
